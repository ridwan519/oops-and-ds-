```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class NodeStack:
    def __init__(self):
        self.top = None

    def is_empty(self):
        return self.top is None

    def push(self, item):
        new_node = Node(item)
        new_node.next = self.top
        self.top = new_node

    def pop(self):
        if not self.is_empty():
            popped_item = self.top.data
            self.top = self.top.next
            return popped_item
        else:
            print("Stack is empty")

    def display_top(self):
        if not self.is_empty():
            print(f"Top item: {self.top.data}")
        else:
            print("Stack is empty")
```

```python

# Example usage:
stack = NodeStack()

```

```python
# Push different data types
stack.push(10)         # Integer
stack.push(3.14)       # Float
stack.push("Hello")    # String
stack.push([1, 2, 3])  # List

```

```python
# Display the top item
stack.display_top()

```

```python
# Pop items and display the top item
print("Popped:", stack.pop())
stack.display_top()

print("Popped:", stack.pop())
stack.display_top()
```
