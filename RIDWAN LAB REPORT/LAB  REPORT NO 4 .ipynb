```python
import math

class Point:
    """Represents a point in two-dimensional geometric coordinates."""
    
    def __init__(self, x: float, y: float) -> None:
        self.x = x
        self.y = y
    
    def distance(self, p2) -> float:
        """Calculates the distance between two points."""
        return math.sqrt((self.x - p2.x)**2 + (self.y - p2.y)**2)
    
    def __str__(self):
        """Returns a string representation of the point."""
        return f"Point({self.x}, {self.y})"


```

```python
# a. Define point1 and pass two numbers.
p1 = Point(3, 4)
p2 = Point(6, 8)

```

```python
# c. Print p1 and p2 using the print command and passing the point as input.
print(p1)  # This will invoke the __str__ method and print Point(3, 4)
print(p2)  # This will invoke the __str__ method and print Point(6, 8)

```

```python
# d. Print the coordinates by using the object and dot operator.
print(f"p1 coordinates: ({p1.x}, {p1.y})")  # Accessing x and y attributes of p1
print(f"p2 coordinates: ({p2.x}, {p2.y})")  # Accessing x and y attributes of p2

```

```python
# e. Print points using the new method (__str__) which is automatically called by print.
# (This is already done in the previous step with print(p1) and print(p2))
```

```python

# f. Calculate the distance between p1 and p2.
distance = p1.distance(p2)
print(f"Distance between p1 and p2: {distance}")
```
