```python
from typing import List, Tuple, Dict

def resistance() -> List[float]:
    return [10.0, 20.0, 30.0, 40.0, 50.0]

def voltage() -> List[float]:
    return [5.0, 10.0, 15.0, 20.0, 25.0]

def current_part_a(resistances: List[float], voltages: List[float]) -> Tuple[List[float], List[float], List[float]]:
    currents = [v / r for v, r in zip(voltages, resistances)]
    return resistances, voltages, currents

def current_part_b(resistances: List[float], voltages: List[float]) -> Dict[str, float]:
    return {f"current_{i+1}": voltages[i] / resistances[i] for i in range(len(resistances))}

```

```python
# Main program
# Call the resistance and voltage functions to get the lists
resistances = resistance()
voltages = voltage()

```

```python
## Part a: Get currents and print all lists in tuple format
resistances_a, voltages_a, currents_a = current_part_a(resistances, voltages)
print("\n--- Part A: Data in Tuple Format ---")
print("Resistances (Ohms):", resistances_a)
print("Voltages (Volts):", voltages_a)
print("Currents (Amperes):", currents_a)

```

```python
# Part b: Get currents and print as a dictionary
currents_b = current_part_b(resistances, voltages)
print("\n--- Part B: Data in Dictionary Format ---")
print("Currents Dictionary:", currents_b)
```

```python
from typing import Union, List, Tuple

def series(resistors: Union[List[float], Tuple[float, ...]]) -> float:
    """
    Calculates the net resistance for resistors connected in series.
    This function can handle any number of resistors provided as a list or tuple.

    :param resistors: A list or tuple of resistances in ohms.
    :return: The net resistance in ohms.
    """
    return sum(resistors)

```

```python
# Print the docstring
print("\n--- Docstring of the series function ---")
print(series.__doc__)

```

```python
# Example usage
resistors = [10.0, 20.0, 30.0, 40.0, 50.0]  # Example list of resistors
net_resistance = series(resistors)

```

```python
# Display the result
print("\n--- Example Output ---")
print(f"Resistors: {resistors}")
print(f"Net Resistance (Series): {net_resistance} Ohms")

```

```python
# Display the function's annotations
print("\n--- Function Annotations ---")
print(series.__annotations__)
```

```python
from typing import Union, List, Tuple

def parallel(resistors: Union[List[float], Tuple[float, ...]]) -> float:
    
    if not resistors:
        raise ValueError("The list of resistors cannot be empty.")
    return 1 / sum(1 / r for r in resistors)

```

```python
# Print the docstring
print("\n--- Docstring of the parallel function ---")
print(parallel.__doc__)

```

```python
# Example usage
resistors = [10.0, 20.0, 30.0, 40.0, 50.0]  # Example list of resistors
net_resistance = parallel(resistors)

```

```python
# Display the result
print("\n--- Example Output ---")
print(f"Resistors: {resistors}")
print(f"Net Resistance (Parallel): {net_resistance:.2f} Ohms")

```

```python
# Display the function's annotations
print("\n--- Function Annotations ---")
print(parallel.__annotations__)
```

```python
from typing import Union, List, Tuple

def display_solar_data(data: Union[List[float], Tuple[float, float, float]]) -> None:
    
    if len(data) != 3:
        raise ValueError("Data must contain exactly three values: [battery_SOC, solar_generation, market_price].")
    
    battery_SOC, solar_generation, market_price = data
    print(f"Battery SOC: {battery_SOC}%")
    print(f"Solar Generation: {solar_generation} kW")
    print(f"Market Price: ₹{market_price} per unit")

```

```python
# Print the docstring
print("\n--- Docstring of the display_solar_data function ---")
print(display_solar_data.__doc__)

```

```python
# Example usage
solar_data = [60, 2, 60]  # Example data in a list
display_solar_data(solar_data)

solar_data_tuple = (60, 2, 60)  # Example data in a tuple
display_solar_data(solar_data_tuple)


```

```python
# Display the function's annotations
print("\n--- Function Annotations ---")
print(display_solar_data.__annotations__)
```

```python
def display_solar_data_dict(*, SOC: float, solar_generation: float, price: float) -> None:
    """
    Displays the battery SOC (%), solar generation (kW), and market price (per unit in rupees)
    from a hybrid solar inverter in a well-formatted manner using dictionary unpacking.

    :param SOC: State of Charge (percentage) of the battery.
    :param solar_generation: Solar power generation in kW.
    :param price: Market price per unit in rupees.
    :return: None
    """
    print(f"Battery SOC: {SOC}%")
    print(f"Solar Generation: {solar_generation} kW")
    print(f"Market Price: ₹{price} per unit")

```

```python
# Print the docstring
print("\n--- Docstring of the display_solar_data_dict function ---")
print(display_solar_data_dict.__doc__)

```

```python
# Example usage with dictionary unpacking
display_solar_data_dict(SOC=60, solar_generation=2, price=60)

# Display the function's annotations
print("\n--- Function Annotations ---")
print(display_solar_data_dict.__annotations__)
```
