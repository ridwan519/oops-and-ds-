```python
# Initial list
my_list = [1, 2, 3]

# a. .append(obj)
my_list.append(4)  # Adds 4 to the end
print(my_list)  # Output: [1, 2, 3, 4]

# b. .extend(iterable)
my_list.extend([5, 6])  # Adds elements from another list
print(my_list)  # Output: [1, 2, 3, 4, 5, 6]

```

```python
# c. .insert(index, obj)
my_list.insert(2, 10)  # Inserts 10 at index 2
print(my_list)  # Output: [1, 2, 10, 3, 4, 5, 6]

# d. .remove(obj)
my_list.remove(10)  # Removes the first occurrence of 10
print(my_list)  # Output: [1, 2, 3, 4, 5, 6]

# e. .pop([index=-1])
my_list.pop()  # Removes and returns the last element
print(my_list)  # Output: [1, 2, 3, 4, 5]
```

```python
# List (mutable)
mutable_list = [1, 2, 3]
mutable_list[1] = 5  # You can change the value of an element
print(mutable_list)  # Output: [1, 5, 3]

# Tuple (immutable)
immutable_tuple = (1, 2, 3)

# Trying to change a tuple element will raise an error
try:
    immutable_tuple[1] = 5  # This will throw an error since tuples are immutable
except TypeError as e:
    print(f"Error: {e}")  # Output: Error: 'tuple' object does not support item assignment
```

```python
num = (1, 2, 3, 4)

# Trying to change an element in a tuple will raise an error
try:
    num[2] = 9  # This will throw an error because tuples are immutable
except TypeError as e:
    print(f"Error: {e}")  # Output: Error: 'tuple' object does not support item assignment
```

```python
# Creating a sample dictionary
my_dict = {
    'name': 'M DANISH KHAN',
    'age': 19,
    'city': 'SWABI',
    'job': 'Engineer'
}
```

```python
# Using .clear()
my_dict.clear()
print(my_dict)  # Output: {}  (an empty dictionary)
```

```python
# Re-creating the dictionary for the demonstration
my_dict = {
    'name': 'M DANISH KHAN',
    'age': 19,
    'city': 'SWABI',
    'job': 'Engineer'
}

# Using .get() to retrieve the value of a specific key
name_value = my_dict.get('name')  # Output: 'Alice'
non_existent_key = my_dict.get('salary')  # Output: None (since 'salary' key doesn't exist)

print(name_value)
print(non_existent_key)
```

```python
# Using .items()
items = my_dict.items()  # Output: dict_items([('name', 'Alice'), ('age', 25), ('city', 'New York'), ('job', 'Engineer')])
print(items)
```

```python
# Using .keys()
keys = my_dict.keys()  # Output: dict_keys(['name', 'age', 'city', 'job'])
print(keys)
```

```python
# Using .values()
values = my_dict.values()  # Output: dict_values(['Alice', 25, 'New York', 'Engineer'])
print(values)
```

```python
squares = []
for x in range(10):
    squares.append(x * x)

print(squares)  # Output: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]
```

```python
squares = [x * x for x in range(10)]
print(squares)  # Output: [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]
```

```python
cubes = [x ** 3 for x in range(10)]
print(cubes)  # Output: [0, 1, 8, 27, 64, 125, 216, 343, 512, 729]
```

```python
even_squares = [x * x for x in range(10) if x % 2 == 0]
print(even_squares)  # Output: [0, 4, 16, 36, 64]
```

```python
even_squares = []
for x in range(10):
    if x % 2 == 0:
        even_squares.append(x * x)

print(even_squares)  # Output: [0, 4, 16, 36, 64]
```

```python
# Define a valid collection, such as a range of numbers
collection = range(10)

# Use list comprehension with a condition
values = [x * x for x in collection if x % 2 == 0]

print(values)  # Output: [0, 4, 16, 36, 64]
```
