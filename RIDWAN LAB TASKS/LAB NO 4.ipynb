```python
# c. Try different names for objects and use different attribute names
# Let's create instances with different names and attributes

obj1 = Point()
obj2 = Point()

```

```python

# Assign attributes with different data types
obj1.x = 10
obj1.y = 20
obj2.name = "DANI"
obj2.age = 19
```

```python
# Print the objects with their attributes
print(obj1)  # Prints the memory address of obj1
print(obj2)  # Prints the memory address of obj2

```

```python
# Print the attributes of the objects
print(f"obj1: x={obj1.x}, y={obj1.y}")
print(f"obj2: name={obj2.name}, age={obj2.age}")
```

```python
# Define the Dog class with the __init__ method
class Dog:
    def __init__(self, name, age):
        self.name = name  # Assign the name to the attribute
        self.age = age    # Assign the age to the attribute

```

```python
# a. Instantiate a Dog object with the name "Miles" and age 4
miles = Dog("Miles", 4)


print(miles)
```
